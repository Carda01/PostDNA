```python
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
from Bio import SeqIO
import random

def generate_random_dna(length):
    return ''.join(random.choices('ACGT', k=length))

def create_fast_file(filename, num_sequences, length):
    records = []
    for i in range(num_sequences):
        seq = generate_random_dna(length)
        record = SeqRecord(Seq(seq), id=f"seq{i}", description="")
        records.append(record)
    SeqIO.write(records, filename, "fasta")
    return filename

# Run the function in a cell
filename = create_fast_file("synthetic_data.fast", 100, 1000)
filename  # Display the generated file name
```

```python
# Read and display the generated FASTA file
def display_fasta_file(filename):
    sequences = []
    for record in SeqIO.parse(filename, "fasta"):
        sequences.append((record.id, str(record.seq)))
    return sequences

# Display the first few sequences from the file
sequences = display_fasta_file(filename)
sequences[:5]  # Show the first 5 sequences
```

```python
import psycopg2
from Bio import SeqIO

conn = psycopg2.connect("dbname=dna user=postgres ")
cur = conn.cursor()

for record in SeqIO.parse("synthetic_data.fast", "fasta"):
    cur.execute("INSERT INTO dna_sequences (name, sequence) VALUES (%s, %s::dna);", (record.id, str(record.seq)))
    
conn.commit()
cur.close()
conn.close()
```
